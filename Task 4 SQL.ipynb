```python
# Import Modules

import sqlite3
```

```python
# Create Connection to Database

connection = sqlite3.connect('books.db')
```

```python
# Import Pandas

import pandas as pd
```

```python
# Viewing the Authors Table's Content

pd.options.display.max_columns = 10
```

```python
# Select everything from authors table

pd.read_sql('SELECT * FROM authors', connection,
            index_col=['id'])
```

```python
# Select everything from titles table

pd.read_sql('SELECT * FROM titles', connection)
```

```python
# Select ISBN table

pd.read_sql('SELECT * FROM author_ISBN', connection)
```

## 1- Select 17.2.2

```python
# Get First and Last Names of Authors

pd.read_sql('SELECT first, last FROM authors', connection)
```

## 2- Where Clause

```python
# Title, Edit, Copyright from Titltes after 2016
pd.read_sql("""SELECT title, edition, copyright
            FROM titles
            WHERE copyright > '2016'""", connection)
```

```python
# Get authors whose last name starts with D

pd.read_sql("""SELECT id, first, last 
            FROM authors
            WHERE last LIKE 'D%'""",
            connection, index_col=['id'])
```

```python
# Get Authors with Specific Pattern

pd.read_sql("""SELECT id, first, last
            FROM authors
            WHERE first LIKE '_b%'""",
            connection, index_col = ['id'])
```

## 3- Order By

```python
# Titles in Ascending Order
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', 
            connection)
```

```python
# Sorting by Multiple Columns Last Name

pd.read_sql("""SELECT id, first, last
            FROM authors
            ORDER BY last, first""",
            connection, index_col=['id'])
```

```python
# Sort authors in descending order by last name and ascending order by first name
pd.read_sql("""SELECT id, first, last
               FROM authors
               ORDER BY last DESC, first ASC""",
            connection, index_col=['id'])
```

```python
# Get ISBN, title, edition, and copyright for titles ending with How To Program, Sorted in ascending order by title

pd.read_sql("""SELECT isbn, title, edition, copyright
            FROM titles
            WHERE title LIKE '%How to Program'
            ORDER BY title""", connection)
```

## 4- Inner Join

```python
# Authors accompanied by ISBNs for books written by each author (Head)

pd.read_sql("""SELECT first, last, isbn
               FROM authors
               INNER JOIN author_ISBN
                   ON authors.id = author_ISBN.id
               ORDER BY last, first""", connection).head()
```
